## Sinal impulso unitário

É o sinal mais simples que existe em tempio discreto. Para representá-lo, usamos a letra $\delta$, e é definido da seguinte maneira:

$$\delta[n] = \left\{\begin{array}{cc}1& ,n = 0 \\ 0& ,c.c \end{array}\right. $$

O seu gráfico é o seguinte:


```python
n = np.arange(-10,11)
x = np.zeros_like(n)
x[n==0] = 1

plt.stem(n,x)
plt.xlabel('n')
plt.ylabel('x');
```

Para obter um impulso unitário em outra posição que não $n=0$, basta atrasarmos (ou adiantarmos) o sinal $\delta[n]$. Se queremos que o sinal seja 1 em $n=k$, basta fazermos $\delta[n-k]$.

---

### Exemplo 1

Por exemplo, se queremos um impulso unitário em $n=2$, fazemos $\delta[n-2]$.

```python
n = np.arange(-10,11)
x = np.zeros_like(n)
x[n==2] = 1

plt.stem(n,x)
plt.xlabel('n')
plt.ylabel('x');
```

## Sinal degrau

Outro sinal que usaremos bastante é o sinal degrau de tempo discreto. Esse sinal é sempre igual a 1 para $n\geq 0$. Para esse sinal usaremos a letra $u$:

$$u[n] =   \left\{\begin{array}{cc}1& ,n \geq 0 \\ 0& ,c.c \end{array}\right. $$

O seu gráfico é mostrado abaixo:

```python
n = np.arange(-10,11)
x = np.zeros_like(n)
x[n>=0] = 1

plt.stem(n,x)
plt.xlabel('n')
plt.ylabel('x');
```

Os sinais degrau e impulso unitário em tempo discreto estão bastante relacionados. O sinal impulso unitário é o sinal degrau subtraido do mesmo sinal degrau atrasado de 1 passo.

$$\delta[n] = u[n] - u[n-1]$$

$\delta[n-2-k]$ é um impulso localizado em $n = 2 + k$. Vamos analisar os primeiros termos dessa sequência de impulsos:

- Quando $k = 2$: O impulso ocorre em $n = 2 + 2 = 4$.
- Quando $k = 3$: O impulso ocorre em $n = 2 + 3 = 5$.
- Quando $k = 4$: O impulso ocorre em $n = 2 + 4 = 6$.
- E assim por diante...

$$x[n] = 1 - \left( \delta[n-4] + \delta[n-5] + \delta[n-6] + \ldots \right)$$

$$x[n] = 1 - \left( u[n-4] - u[n-5] + u[n-5] - u[n-6] + u[n-6] - \ldots \right)$$

$$x[n] = 1 - \left( u[n-4] + u[n-5] + u[n-6] + \ldots \right)$$

A série infinita de degraus $u[n-4] + u[n-5] + u[n-6] + \ldots$ resulta em um degrau unitário no ponto $n = 4$. Isso porque, a partir de $n = 4$, todos os termos subsequentes serão iguais a 1.

$$x[n] = 1 - u[n-4]$$

A forma proposta é $x[n] = u[Mn - n_0]$. Para isso, $M$ deve ser 1 (pois os impulsos estão separados por 1 unidade de tempo) e $n_0$ deve ser 2 (pois o primeiro impulso ocorre em $n = 4$, que é $2 + 2$).

$$x[n] = u[1 \cdot n - 2]$$

$$x[n] = u[n - 2]$$

$$-1 < n < b \quad \text{(devido a } u[n+2] \text{)}$$
$$n \geq 8 \quad \text{(devido a } u[n-8] \text{)}$$

$$a = 7 \quad \text{e} \quad b = 8$$
